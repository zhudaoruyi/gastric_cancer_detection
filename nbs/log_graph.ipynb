## g,data augumentation include random rotate ,random flip ,random hue jitter

```python
import numpy as np
import matplotlib.pyplot as plt
# plt.rcParams['font.sans-serif']=['SimHei'] #
# plt.rcParams['font.family']='sans-serif'
# plt.rcParams['axes.unicode_minus']=False #
%matplotlib inline

train_log = open("log10032100_01.log")
log_line = train_log.readlines()
epoch, acc, loss, val_acc, val_loss = [], [], [], [], []
for i in range(len(log_line)):
    if i<len(log_line)-1:
        line_list = list(eval(log_line[i+1]))
        epoch.append(line_list[0])
        acc.append(line_list[1])
        loss.append(line_list[2])
        val_acc.append(line_list[3])
        val_loss.append(line_list[4])
fig, ax = plt.subplots(1,1,figsize=(10, 10))    # 
# ax = axes.ravel()
ax.set_title('acc-loss curve')
ax.plot(epoch,acc,color='green',linewidth=2.5,label='train acc')
ax.plot(epoch,loss,color='green',linewidth=2.5,label='train loss')
ax.plot(epoch,val_acc,color='red',linewidth=1.5,label='valid acc')
ax.plot(epoch,val_loss,color='red',linewidth=1.5,label='valid loss')
ax.set_yticks([0.0,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1.0])
ax.legend(loc=7)
```
